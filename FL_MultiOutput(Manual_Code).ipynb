```python
def custom_loss_hyper(y_true, y_pred,alpha = 1): 
    diff = y_true - y_pred
    
    multi = alpha * K.switch(K.abs(diff) < 1,(K.abs(diff)+1),K.abs(diff))    
    nmse = K.mean(K.switch(K.greater_equal(y_true,70),(K.square(y_pred-y_true)+(multi*K.pow((K.abs(y_true-125)),2))),(K.square(y_pred-y_true)+K.pow((y_true-125),2))))
    return nmse
```

```python
def custom_loss_hypo(y_true, y_pred,alpha = 1): 
    diff = y_true - y_pred
    
    multi = alpha * K.switch(K.abs(diff) < 1,(K.abs(diff)+1),K.abs(diff))    
    nmse = K.mean(K.switch(K.greater_equal(y_true,70),(K.square(y_pred-y_true)+K.pow((y_true-125),2)),
                                                      (K.square(y_pred-y_true)+(multi*K.pow((K.abs(y_true-125)),2)))))
    return nmse
```

```python
# load models from file
def load_all_models(weights_list):
    all_models = list()
    weights_list = list(weights_list.values())
    for i in range(len(weights_list)):
        
        ## LSTM Model
        n_timesteps, n_features, n_outputs = 24, 1, 1
        # define model
        model = create_keras_model()
        model.compile(
                    loss = {
                        "hypo_loss": custom_loss_hypo,
                        "hyper_loss": custom_loss_hyper
                    },

                    metrics = {
                        "hypo_loss": [tf.keras.metrics.MeanSquaredError()],
                        "hyper_loss": [tf.keras.metrics.MeanSquaredError()]
                    },

                    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
                    ) 
        # load model weights from list
        Model = clone_model(model)
        Model.set_weights(weights_list[i])
        # add to list of members
        all_models.append(Model)
        
    return all_models   
```

```python
# create a model from the weights of multiple models
def model_weight_ensemble(members, weights):
	# determine how many layers need to be averaged
	n_layers = len(members[0].get_weights())
	# create an set of average model weights
	avg_model_weights = list()
	for layer in range(n_layers):
		# collect this layer from each model
		layer_weights = array([model.get_weights()[layer] for model in members])
		# weighted average of weights for this layer
		avg_layer_weights = average(layer_weights, axis=0, weights=weights)
		# store average layer weights
		avg_model_weights.append(avg_layer_weights)
	# create a new model with the same structure
	model = clone_model(members[0])
	# set the weights in the new
	model.set_weights(avg_model_weights)
# 	model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
	model.compile(
                    loss = {
                        "hypo_loss": custom_loss_hypo,
                        "hyper_loss": custom_loss_hyper
                    },

                    metrics = {
                        "hypo_loss": [tf.keras.metrics.MeanSquaredError()],
                        "hyper_loss": [tf.keras.metrics.MeanSquaredError()]
                    },

                    optimizer = opt_s
                    ) 
	return model
```
